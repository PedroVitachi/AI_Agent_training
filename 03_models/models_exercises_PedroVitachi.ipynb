# Regression, Classification, Decision Trees, and Ensemble Models


## 1. Regressão

Na regressão linear, a multicolinearidade entre as variáveis independentes pode desestabilizar os coeficientes do modelo. Explique como a **Regressão Ridge** e a **Regressão Lasso** abordam essa questão de maneiras diferentes. Em quais cenários você preferiria uma em vez da outra?

R: A regressão Lasso trata a multicolinearidade ao pegar as features com uma correlação mais alta, selecionar uma delas para manter seu peso na fórmula regressão, e remover o peso das outras features, basicamente limpando um pouco o modelo de features similares que podem pesar a analise e aumentar o bias.

Na regressão ridge, essa tratativa é diferente, os coeficientes são elevados ao quadrado, expondo melhor as features que tem uma correlação maior, sem necessáriamente remover uma feature da equação.

Sobre qual desses modelos utilizar, vai depender de como as features serão tratadas, se uma das intenções é focar na simplicidade e facilidade de interpretação do modelo, a regressão lasso seria mais recomendada, já a ridge seria utilizada quando se deseja manter as features, mesmo que tenham suas variâncias, trazendo uma analise mais detalhada.

## 2. Classificação

Explique como o **desequilíbrio de classes** afeta o desempenho dos modelos de classificação. Quais estratégias você usaria para lidar com um conjunto de dados com uma variável alvo altamente desbalanceada, e como essas estratégias impactariam as métricas de avaliação do modelo?

R: O desequilíbrio pode ser observado quando as amostragens de dado tendem a ter uma tendência maior para uma das amostragem, gerando um viés para a amostragem majoritária, sendo assim, o modelo seria muito preciso para essa amostragem, porém ao ser utilizado em outra amostragem que não possua a mesma tendência nos dados ele terá um resultado muito inferior.

Uma das maneiras que podem ser utilizadas para evitar esse desequilíbrio seria utilizar uma fatia menor de dados da classe majoritária, poderia também utilizar mais dados da classe minoritária ou até os dois juntos, ou partir para utilização de alguns métodos que vão selecionar aleatoriamente as fatias de dados das duas classes.

## 3. Árvores de Decisão

Árvores de Decisão são conhecidas pela tendência de se ajustarem demais aos dados de treinamento. Descreva o papel da **poda de árvores** e do **hiperparâmetro de profundidade máxima** no controle do overfitting. Como essas técnicas equilibram viés e variância no modelo?

R: Um dos problemas das árvores de decisão é o fato que a análise pode se estender bastante, causando um certo enviesamento ao focar muito nos dados específicos em vez de tratar os dados de uma maneira geral, basicamente "decorando" os valores das features do dataset e gerando previsões que só seriam precisas com aquele conjunto de dados.

Na poda, durante o treinamento do modelo (ou após, caso use a metodologia da pós poda) o crescimento da árvore pode ser interrompido com base em vários parametros que visam diminuir a complexidade da árvore, na pós poda isso é feito após a geração completa da árvore.

Ao utilizar Hiperparâmetro de profundidade máxima, o crescimento da árvore é limitado antes mesmo da sua geração, sendo informado o tamanho máximo da profundidade, dependendo da complexidade dos dados, evitando bem o over fitting que acontece em árvores mais complexas.

## 4. Modelos de Ensemble

Compare e contraste **Bagging** (por exemplo, Random Forest) e **Boosting** (por exemplo, Gradient Boosting Machines) em termos de suas abordagens para reduzir variância e viés. Como cada método afeta a robustez e interpretabilidade do modelo final?

R: A maior diferença entre Bagging e Boosting se encontra na maneira em que os dois aplicam os modelos de treinamento. No Bagging, o dataset é separado (geralmente de maneira aleatória e com reposição) para que cada modelo que for aplicado tenha uma base diferente e uma previsão diferente, e no final, suas previsões são testadas no mesmo dataset de teste, calculando uma média dos modelos.

Já o Boosting trabalha de maneira diferente, os modelos são aplicados de maneira sequencial, visando utilizar os resultados dos modelos anteriores, aumentando em si a precisão da análise, porém pode causar alguns problemas de overfitting caso existam muitos modelos sendo executados, já que os dados ficariam muito tendenciosos.

## 5. Conjuntos de Árvores para Regressão vs Classificação

Métodos de ensemble baseados em árvores, como Random Forests e Gradient Boosting, podem ser usados tanto para regressão quanto para classificação. Descreva as principais diferenças em como esses algoritmos são configurados e avaliados para problemas de regressão e classificação. Como os critérios de divisão e as métricas de desempenho mudam entre esses contextos?

R: A grande diferença se encontra em como são avaliadas a regressão e a classificação, na classificação, visamos gerar uma predição de um valor categórico, e não um valor numérico, como é visto na regressão.

na CLassificação, a divisão de métricas é feita de maneira em que os nós estão relacionados, visando avaliar a desordem ou proporção das obervações contidas naquele nó, e geralmente suas métricas de desempenho são tratados com medidas de precisão, focadas no número de acerto por previsões, entre outros.

No caso da Regressão, como estamos prevendo dados numéricos, tanto os critérios de divisão quanto as métricas circulam em volta de tratar a quantidade de erro, utilizando métodos como MAE, RMSE e R2
